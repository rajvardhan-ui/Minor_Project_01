```python
import os

# Feature_01 -The chat parser
def parse_whatsapp_chat(filename):
  parsed_messages = []
  unique_days = set()
  system_count = 0
  media_count = 0
  deleted_count = 0

  # Check if the file exists
  if not os.path.exists(filename):
    print(f"Error: File '{filename}' not found.")
    return [], set() # Return empty list and set if file not found

  with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()                                                 # Opening file
  for line in lines:
    line = line.strip()

    if not line:
      continue

    if " - " in line:
      parts = line.split(" - ",1)
      timestamp = parts[0]
      rest = parts[1]

      date_part = timestamp.split(",")[0].strip()
      unique_days.add(date_part)


      if ": " in rest:
        sender, text = rest.split(": ",1)


        if text == "<Media omitted>":
          media_count += 1
        elif text == "This message was deleted":
          deleted_count += 1


        parsed_messages.append({
           'timestamp': timestamp,
            'sender': sender,
            'text': text
        })
      else:
        system_count+=1
    else:
      pass

  print("Feature_01 -The chat parser")
  print(f"Total Messages: {len(parsed_messages)}")
  print(f"Unique Days: {len(unique_days)}")
  print(f"System Messages: {system_count}")
  print(f"Media Messages: {media_count}")
  print(f"Deleted Messages: {deleted_count}")

  return parsed_messages, unique_days

chat_data, unique_data_set = parse_whatsapp_chat('hostel_bois.txt')
```

```python
# Feature_02: Group Overview


def group_overview(parsed_messages, unique_data_set):
  msg_count = {}
  word_counts = {}
  total_charts = {}

  for msg in parsed_messages:
    sender = msg['sender']
    text = msg['text']

    words = text.split()
    num_words = len(words)
    num_chars = len(text)

    if sender not in msg_count:
      msg_count[sender] = 0
      word_counts[sender] = 0
      total_charts[sender] = 0

    msg_count[sender] += 1
    word_counts[sender] += num_words
    total_charts[sender] += num_chars

  total_real_msgs = len(parsed_messages)
  total_days = len(unique_data_set)
  total_participants = len(msg_count)

  sorted_participants = sorted(msg_count.items(), key=lambda item: item[1], reverse=True)


  print("="*80)
  print('"GROUPDNA REPORT - "Hostel Bois 4ever"')
  print(f"{total_days} days • {total_real_msgs} messages • {total_participants} members")
  print("="*80)

  print(f"Total Messages   : {total_real_msgs}")
  print(f"Total Participants: {total_participants}")
  print("\nMessages per person")


  for sender, count in sorted_participants:
    percentage = (count / total_real_msgs) * 100

    bar_length = int(percentage / 2)
    bar_visual = "█" * bar_length
    print(f" {sender:<12} {bar_visual:<20} {count:>3} ({percentage:.1f}%)")

  print("\n"+"-"*80)
  print("PER-PERSON TEXT METRICS")
  print("-"*80)
  for sender, count in sorted_participants:
    w_count = word_counts[sender]
    avg_len = total_charts[sender] / count if count > 0 else 0
    print(f"User: {sender:<12} | Total Words: {w_count:<5} | Avg Msg Length: {avg_len:.1f} chars")

group_overview(chat_data, unique_data_set)
```

```python
# Feature_03: Most active Day and Hour

def analyze_activity_peaks(parsed_messages):
    day_counts = {}
    hour_counts = {}

    for msg in parsed_messages:
        timestamp = msg['timestamp']

        date_part, time_part = timestamp.split(", ")
        hour_part = time_part.split(":")[0]


        if date_part not in day_counts:
            day_counts[date_part] = 0
        day_counts[date_part] +=1

        if hour_part not in hour_counts:
            hour_counts[hour_part] = 0
        hour_counts[hour_part] += 1

    best_day, max_day_msgs = max(day_counts.items(), key=lambda X: X[1])
    best_hour, max_hour_msgs = max(hour_counts.items(), key=lambda X: X[1])


    hour_int = int(best_hour)
    hour_window = f"{hour_int:02d}:00 - {hour_int+1:02d}:00"

    avg_msgs_per_day = max_hour_msgs / 60



    print("\n"+"="*80)
    print("ACTIVITY PEAKS")
    print("="*80)
    print(f"Most active day: {best_day} with {max_day_msgs} messages")
    print(f"Most active hour: {hour_window} with {max_hour_msgs} messages")
    print("\n"+"="*80)

analyze_activity_peaks(chat_data)

```

```python
# Feature_04 : Activity Heatmap
import numpy as np
def activity_heatmap(parsed_messages):
  participants = ['Rahul','Priya','Aman','Karan','Neha','Vikas']
  user_to_idx = {name: idx for idx, name in enumerate(participants)}

  heatmap_matrix = np.zeros((6,24))


  for msg in parsed_messages:
      sender = msg['sender']
      if sender in user_to_idx:
        user_idx = user_to_idx[sender]
        date_part, time_part = msg['timestamp'].split(",")
        hour_idx = int(time_part.split(":")[0])
        heatmap_matrix[user_idx, hour_idx] += 1

  print("\n"+"="*80)
  print("ACTIVITY HEATMAP")
  print("="*80)
  print("             00 03  06  09  12  15  18  21 24")


  for idx, name in enumerate(participants):
      row_values = heatmap_matrix[idx]
      max_val = np.max(row_values) if np.max(row_values) > 0 else 1

      shaded_line = " "

      for hour in range(24):
        val = row_values[hour]
        percentage = (val / max_val) * 100

        if percentage <= 25:
          shade = "."
        elif percentage <= 25:
          shade = '░'
        elif percentage <= 50:
          shade = "▒"
        else:
          shade = '█'

        if hour in [0,3,6,9,12,15,18,21]:
            shaded_line += shade + " "
        else:
            shaded_line += shade

      print(f"{name:<11} {shaded_line}")

  print("\n"+"="*80)



activity_heatmap(chat_data)

```

```python
# Feature_05 : Word Frequency Analysis
def word_frequency_analysis(parsed_messages):
    word_frequency = {}

    stop_words = {
        'i', 'me', 'my', 'we', 'our', 'you', 'your', 'he', 'him', 'his', 'she',
        'her', 'it', 'they', 'them', 'that', 'this', 'these', 'those', 'am', 'is',
        'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do',
        'does', 'did', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as',
        'of', 'at', 'by', 'for', 'with', 'about', 'to', 'in', 'on', 'no', 'not',
        'hai', 'kya', 'bhi', 'se', 'ko', 'hi', 'ye', 'toh', 'kuch', 'hoga'
    }
    punctuation_chars = '.,!?;:"()[]{}-_#@*\\/<>='


    for msg in parsed_messages:
      text_lower = msg['text'].lower()

      cleaned_text = " "
      for char in text_lower:
        if char not in punctuation_chars:
          cleaned_text += char
        else:
          cleaned_text += " "

      words = cleaned_text.split()
      for word in words:
        word = word.strip()
        if word and word not in stop_words:
          word_frequency[word] = word_frequency.get(word, 0) + 1

    sorted_words = (sorted(word_frequency.items(), key=lambda item: item[1], reverse=True))
    top_10 = sorted_words[:10]


    print("\n"+"="*80)
    print("THIS GROUPS FAVOURITE WORDS")
    print("="*60)

    for index, (word,count) in enumerate(top_10, 1):
      bar_visual = "█"*(count // 10 if count>10 else count)
      print(f" #{index:<2} │ {word:<12} │ {bar_visual} {count}")

    print("="*80)

word_frequency_analysis(chat_data)


```

```python
# Feature_06 : Responce Sheet & Silent Streaks

from datetime import datetime

def analyze_behaviour_metrics(parsed_messages, unique_data_set):
    participants = ['Rahul','Priya','Aman','Karan','Neha','Vikas']


    total_reply_secs = {name: 0 for name in participants}
    reply_counts = {name: 0 for name in participants}

    for i in range (1, len(parsed_messages)):
      prev_msg = parsed_messages[i-1]
      curr_msg = parsed_messages[i]

      prev_sender = prev_msg['sender']
      curr_sender = curr_msg['sender']

      if prev_sender != curr_sender and curr_sender in participants:
        time_format = "%d/%m/%y, %H:%M"
        prev_timestamp = datetime.strptime(prev_msg['timestamp'], time_format)
        curr_timestamp = datetime.strptime(curr_msg['timestamp'], time_format)

        # Corrected: Use datetime objects for .date()
        if prev_timestamp.date() == curr_timestamp.date():
          time_delta = curr_timestamp - prev_timestamp
          seconds_gap = time_delta.total_seconds()

          if 0 <= seconds_gap <= 120:
            total_reply_secs[curr_sender] += seconds_gap
            reply_counts[curr_sender] += 1


    date_format = "%d/%m/%y"
    # Corrected: date_str is already a string, no need to call it as a function
    sorted_dates = sorted(list(unique_data_set), key=lambda date_str: datetime.strptime(date_str, date_format))

    daily_activity = {name: {dt: 0 for dt in sorted_dates}for name in participants}
    for msg in parsed_messages:
      sender = msg['sender']
      if sender in participants:
            # Corrected: .split(',') returns a list, need to get the first element before stripping
            date_part = msg['timestamp'].split(",")[0].strip()
            if date_part in daily_activity[sender]:
              daily_activity[sender][date_part] += 1

    longest_silent_streaks = {name: 0 for name in participants}

    for name in participants:
      current_streak = 0
      max_streak = 0
      for dt in sorted_dates:
        if daily_activity[name][dt] == 0:
          current_streak += 1
          max_streak = max(max_streak, current_streak)
        else:
          current_streak = 0

    longest_silent_streaks[name] = max_streak

    print("\n"+"="*80)
    print("BEHAVIOUR METRICS")
    print("="*80)

    print(f" {'User':<12} | {'Avg Responce Speed':<22} | {'Longest Silent Streak':<22}")
    print("-"*80)

    for name in participants:
      avg_sec = total_reply_secs[name]
      counts = reply_counts[name]
      avg_min = (avg_sec / counts)/ 60 if counts > 0 else 0

      streak_days = longest_silent_streaks[name]

      # Corrected: f-string formatting typo
      speed_display = f"{avg_min:.1f}mins({counts} replies)" if counts > 0 else "N/A"
      streak_display = f"{streak_days} days" if streak_days > 0 else "N/A"

      print(f" {name:<11} | {speed_display:<22} | {streak_display:<22}")

    print("\n"+"="*80)

    return total_reply_secs, longest_silent_streaks

analyze_behaviour_metrics(chat_data, unique_data_set)
```

```python
# Feature_07 : Personality Archtype Detection Engine

# 1. THE SPAMMER:
def score_spammer(user_msgs, total_user_count, total_groupmsgs, total_days):
    if not user_msgs: return 0
    burst_count = 0
    total_burst_count = 0
    in_burst = False
    current_length = 0

    short_msgs = sum(1 for m in user_msgs if len(m['text']) < 15)
    return (short_msgs / total_user_count) * 100 if total_user_count > 0 else 0

# 2. THE GROP MOM
def score_group_mom(user_msgs, total_user_count, total_group_count, total_days):
  care_words = ['okay', 'safe', 'eat', 'sleep', 'take care', 'please', 'tc', 'dhyaan']
  match_count = 0
  for m in user_msgs:
    txt = m['text'].lower()
    match_count += sum(1 for word in care_words if word in txt)
  return (match_count / total_user_count) * 100 if total_user_count > 0 else 0

# 3. THE NIGHT OWL
def score_night_owl(user_msgs, total_user_count, total_group_msgs, total_days):
  night_count = 0
  for m in user_msgs:

    time_part = m['timestamp'].split(", ")[1] # Get 'HH:MM' part, e.g., '01:17'
    hour = int(time_part.split(':')[0])       # Get 'HH' and convert to int, e.g., 1

    if hour >= 22 or hour <= 6:
      night_count += 1
  return (night_count / total_user_count) * 100 if total_user_count >0 else 0

# 4. THHE STORYTELLER
def score_storyteller(user_msgs, total_user_count, total_groupmsgs, total_days):
  if total_user_count == 0: return 0
  total_words = sum(len(m['text'].split()) for m in user_msgs)
  avg_words = total_words / total_user_count
  return avg_words * 2

# 5. THE DRAMA QUEEN
def score_drama_queen(user_msgs, total_user_count, total_groupmsgs, total_days):
  drama_count = 0
  for m in user_msgs:
    txt = m['text']
    if '😂' in txt:
      drama_count += 1
  return (drama_count / total_user_count) * 100 if total_user_count > 0 else 0

# 6. THE GHOST
def score_ghost(user_msgs, total_user_count, total_groupmsgs, total_days):
  active_days = set()
  for m in user_msgs:
    active_days.add(m['timestamp'].split(',')[0])
  return (len(active_days) / total_days) * 100 if total_days > 0 else 0

# 7. THE COMEDIAN
def score_comedian(user_msgs, total_user_count, total_groupmsgs, total_days):
  slang_words = ['wtf', 'lol', 'lmao', 'haha', 'rofl', 'xd', 'funny']
  match_count = 0
  for m in user_msgs:
    txt = m['text'].lower()
    match_count += sum(1 for word in slang_words if word in txt)
  return (match_count / total_user_count) * 100 if total_user_count > 0 else 0

# 8. THE QUESTION MASTER
def score_question_master(user_msgs, total_user_count, total_groupmsgs, total_days):
  q_count = sum(1 for m in user_msgs if m['text'].strip().endswith('?'))
  return (q_count / total_user_count) * 100 if total_user_count > 0 else 0

# ---ORCHASTRIAN PROFILE ENGINE---
def detect_group_archtypes(parsed_messages, total_days = 60):
  participants = ['Rahul','Priya','Aman','Karan','Neha','Vikas']
  archtype_labels = [
      'The Spammer', 'The Group Mom', 'The Night Owl', 'The Storyteller',
      'The Drama Queen', 'The Ghost', 'The Comedian', 'The Question Master'
  ]
  assigned_profiles = {}
  total_groupmsgs = len(parsed_messages)

  for name in participants:
    user_msgs = [m for m in parsed_messages if m['sender'] == name]
    total_user_count = len(user_msgs)

    scores = [
        score_spammer(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_group_mom(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_night_owl(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_storyteller(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_drama_queen(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_ghost(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_comedian(user_msgs, total_user_count, total_groupmsgs, total_days),
        score_question_master(user_msgs, total_user_count, total_groupmsgs, total_days)
    ]

    max_idx = scores.index(max(scores)) if total_user_count > 0 else 5
    assigned_profiles[name] = archtype_labels[max_idx]

  print("\n"+"="*80)
  print("GROUP ARCHTYPE DETECTION")
  print("="*80)
  for name, role in assigned_profiles.items():
    print(f"  {name:<12} ➔ Assigned Archetype: {role}")
  print("\n"+"="*80)

detect_group_archtypes(chat_data)
```
